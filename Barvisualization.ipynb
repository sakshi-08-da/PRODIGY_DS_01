```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Set the plotting style
plt.style.use('bmh')

# Load the dataset
path="/content/countries-table.csv"
df = pd.read_csv(path)
df.head(10)

# Sort the dataframe by growthRate in descending order and get top 10 countries
top_df = df.sort_values(by='growthRate', ascending=False).head(10)

# Extract data for plotting
x = top_df['growthRate']
y = top_df['country']

# Create a horizontal bar chart
plt.barh(y, x, color='skyblue' )
plt.xlabel('Growth Rate')
plt.ylabel('Country')
plt.title('Top 10 Countries with Highest Growth Rate')
plt.gca().invert_yaxis()
plt.tight_layout
plt.show()
```
